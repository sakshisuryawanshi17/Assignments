## Q.3 Sales of products in four different regions is tabulated for males and females. Find if male-female buyer rations are similar across regions.
1.	Check p-value
2.	If p-Value < alpha, we reject Null Hypothesis


```python
BuyerRatio=pd.read_csv("/content/BuyerRatio Hypo.csv")
BuyerRatio
```

```python
BuyerRatio.describe()
```

```python
BuyerRatio.info()
```

```python
df_table=BuyerRatio.iloc[:,1:6]
df_table
```

```python
df_table.values
```

```python
val=stats.chi2_contingency(df_table)
val
```

```python
type(val)
```

```python
no_of_rows=len(df_table.iloc[0:2,0])
no_of_columns=len(df_table.iloc[0,0:4])
degree_of_f=(no_of_rows-1)*(no_of_columns-1)
print('Degree of Freedom=',degree_of_f)
```

```python
Expected_value=val[3]
Expected_value
```

```python
from scipy.stats import chi2
```

```python
chi_square=sum([(o-e)**2/e for o,e in zip(df_table.values,Expected_value)])
chi_square_statestic=chi_square[0]+chi_square[1]
chi_square_statestic
```

```python
critical_value=chi2.ppf(0.95,3)
critical_value
```

```python
if chi_square_statestic >= critical_value:
	print('Dependent (reject H0)')
else:
	print('Independent (fail to reject H0)')
```

```python
pvalue=1-chi2.cdf(chi_square_statestic,3)
pvalue
```

```python
if pvalue <= 0.05:
	print('Dependent (reject H0)')
else:
	print('Independent (fail to reject H0)')
```
