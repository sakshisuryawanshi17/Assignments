```python
#Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import string
import spacy

from matplotlib.pyplot import imread
from wordcloud import WordCloud,STOPWORDS
%matplotlib inline

from nltk.tokenize import TweetTokenizer
```

```python
tweet=pd.read_csv('/content/Elon_musk Test Mining.csv',encoding='latin-1')
tweet
```

```python
tweet.drop(['Unnamed: 0'],inplace=True,axis=1)
```

```python
tweet
```

```python
tweet=[Text.strip() for Text in tweet.Text]
```

```python
tweet=[Text for Text in tweet if Text]
```

```python
tweet[0:10]
```

```python
# Joining the list into one string/text
tweet_text=' '.join(tweet)
tweet_text
```

```python
tknzr = TweetTokenizer(strip_handles=True)
tweet_tokens=tknzr.tokenize(tweet_text)
print(tweet_tokens)
```

```python
# Again Joining the list into one string/text
tweet_tokens_text=' '.join(tweet_tokens)
tweet_tokens_text
```

```python
# Remove Punctuations
no_punc_text=tweet_tokens_text.translate(str.maketrans('','',string.punctuation))
no_punc_text
```

```python
# remove https or url within text
import re
no_url_text=re.sub(r'http\S+','',no_punc_text)
no_url_text
```

```python
# Tokenization
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
from nltk.tokenize import word_tokenize
text_tokens=word_tokenize(no_url_text)
print(text_tokens)
```

```python
# Tokens count
len(tweet_tokens)
```

```python
#Remove Stopwords
from nltk.corpus import stopwords
my_stop_words=stopwords.words('english')

sw_list=['\x92','rt','ye','yeah','haha','Yes','U0001F923','I']
my_stop_words.extend(sw_list)

no_stop_token=[word for word in text_tokens if not word in my_stop_words]
print(no_stop_token)
```

```python
#Normalize the data
lower_words=[Text.lower() for Text in no_stop_token]
print(lower_words[100:200])
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
stemmed_token=[ps.stem(word)for word in lower_words]
print(stemmed_token[100:200])
```

```python
# Lemmatization
nlp=spacy.load('en_core_web_sm')
doc=nlp(' '.join(lower_words))
print(doc)
```

```python
lemmas=[token.lemma_ for token in doc]
print(lemmas)
```

```python
clean_tweet=' '.join(lemmas)
clean_tweet
```

## Generate word cloud

```python
# Define a function to plot word cloud
def plot_cloud(wordcloud):
    plt.figure(figsize=(40,30))
    plt.imshow(wordcloud)
    plt.axis('off')

# Generate Word Cloud

STOPWORDS.add('pron')
STOPWORDS.add('rt')
STOPWORDS.add('yeah')
wordcloud=WordCloud(width=3000,height=2000,background_color='black',max_words=50,
                   colormap='Set1',stopwords=STOPWORDS).generate(clean_tweet)
plot_cloud(wordcloud)
```

## Named Entity Recognition (NER)



```python
# Parts Of Speech (POS) Tagging
nlp=spacy.load('en_core_web_sm')

one_block=clean_tweet
doc_block=nlp(one_block)
spacy.displacy.render(doc_block,style='ent',jupyter=True)
```

```python
for token in doc_block[100:200]:
    print(token,token.pos_)
```

```python
# Filtering the nouns and verbs only
nouns_verbs=[token.text for token in doc_block if token.pos_ in ('NOUN','VERB')]
print(nouns_verbs[100:200])
```

```python
# Counting the noun & verb tokens
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()

X=cv.fit_transform(nouns_verbs)
sum_words=X.sum(axis=0)

words_freq=[(word,sum_words[0,idx]) for word,idx in cv.vocabulary_.items()]
words_freq=sorted(words_freq, key=lambda x: x[1], reverse=True)

wd_df=pd.DataFrame(words_freq)
wd_df.columns=['word','count']
wd_df[0:10]
# top ten results
```

```python
# Visualization
wd_df[0:10].plot.bar(x='word',figsize=(12,8),title='Top 10 nouns and verbs')
plt.show()
```

## Emotion Mining - Sentiment Analysis

```python
from nltk import tokenize
sentences=tokenize.sent_tokenize(' '.join(tweet))
sentences
```

```python
sent_df=pd.DataFrame(sentences,columns=['sentence'])
sent_df
```

```python
affin=pd.read_csv('/content/Afinn.csv',sep=',',encoding='Latin-1')
affin
```

```python
affinity_scores=affin.set_index('word')['value'].to_dict()
affinity_scores
```

```python
nlp=spacy.load('en_core_web_sm')
sentiment_lexicon=affinity_scores

def calculate_sentiment(text:str=None):
    sent_score=0
    if text:
        sentence=nlp(text)
        for word in sentence:
            sent_score+=sentiment_lexicon.get(word.lemma_,0)
    return sent_score
# manual testing
calculate_sentiment(text='great')
```

```python
sent_df['sentiment_value']=sent_df['sentence'].apply(calculate_sentiment)
sent_df['sentiment_value']
```

```python
sent_df['word_count']=sent_df['sentence'].str.split().apply(len)
sent_df['word_count']
```
