## Kmeans

```python
import pandas as pd
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
```

```python
crim=pd.read_csv('/content/crime_data Clustering.csv')
crim
```

```python
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)
```

```python
df_norm=norm_func(crim.iloc[:,1:])
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(df_norm)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel("WCSS")
plt.show()
```

```python
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crim['Clust']=x
```

```python
crim
```

```python
crim.iloc[:,1:5].groupby(crim.Clust).mean()
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crim=pd.read_csv('/content/crime_data Clustering.csv')
crim
```

```python
crim.info()
```

```python
df=crim.iloc[:,1:5]
df
```

```python
df.values
```

```python
stscaler=StandardScaler().fit(df.values)
x=stscaler.transform(df.values)
```

```python
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
airline=pd.read_csv('/content/EastWestAirlines.csv')
airline
```

```python
airline.head()
```

```python
airline.info()
```

```python
airline2=airline.drop(['ID'],axis=1)
airline2
```

```python
airline2_norm=pd.DataFrame(normalize(airline2),columns=airline2.columns)
airline2_norm
```

```python
plt.figure(figsize=(10,7))
dendrograms=sch.dendrogram(sch.linkage(airline2_norm,'complete'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(airline2_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
airline2['clustersid']=hclusters.labels_
airline2
```

```python
airline2.groupby('clustersid').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(airline2['clustersid'],airline2['Balance'], c=hclusters.labels_)
```
